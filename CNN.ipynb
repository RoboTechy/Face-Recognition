# import necessary libraries

```python
import tensorflow as tf
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Load pre-trained model (e.g., MTCNN for face detection)

```python
from mtcnn.mtcnn import MTCNN
detector = MTCNN()
```

# Read the input image

```python
image = cv2.imread('Personal.jpg')
```

# Detect faces

```python
faces = detector.detect_faces(image)
```

# Draw rectangles around the faces

```python
for face in faces:
    x, y, w, h = face['box']
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

# Display the output

```python
plt.imshow(image[:,:,::-1])
```
